```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.DataFrame([(1,"Goroka Airport","Goroka","Papua New Guinea","GKA","AYGA",-6.081689834590001,145.391998291,5282,10,"U","Pacific/Port_Moresby","airport","OurAirports")], 
                  columns=["Airport ID", "Name", "City", "Country", "IATA", "ICAO", "Latitude", "Longitude", 
                           "Altitude", "Timezone", "DST", "Tz database time zone", "Type", "Source"])
```

```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
file = open('airports.dat','r')
for i, line in enumerate(file):
   try:
      a = line[:-1].replace('"', '')
      #print(line)
      df.loc[i] = a.split(',')
   except ValueError:
      df.loc[i] = [i.replace('"', '') for i in re.findall(r'".*?"|[\d|\.|\\\\N]+', line)]

file.close()
```

```python
df = df.astype({'Airport ID': int, "Latitude":float, "Longitude":float, "Altitude":int})
```
